```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

titanic = pd.read_csv('/content/sample_data/train.csv')
```

image.png

```python
x1 = titanic['Age']
y1 = titanic['Fare']
plt.scatter(x1, y1, c ="red",
			linewidths = 1,
			marker ="*",
			edgecolor ="blue",
			s = 200)
plt.title("Scatter Plot")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

image.png

```python
plt.scatter(titanic['Age'], titanic['Fare'], c=titanic['Pclass'],s=titanic['Pclass'])
plt.title("Scatter Plot")
plt.xlabel('Age')
plt.ylabel('Fare')
plt.colorbar()
plt.legend()
plt.show()
```

image.png

```python
titan=titanic.head(20)
plt.plot(titan['Age'],marker='o')
plt.plot(titan['Fare'],marker='o')
plt.title("Line Chart")
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```

```python
plt.hist(titanic['Age'], color = "skyblue", ec="blue")
plt.title("Histogram")
plt.show()
```

```python
sns.boxplot(x='Age', y="Sex",data=titanic );
plt.show()
```
